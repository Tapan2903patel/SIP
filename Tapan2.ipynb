```python
from matplotlib import pyplot as plt
import numpy as np
fig,ax=plt.subplots()
x=np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

ax.annotate('local maximum',xy=(6.28,1),xytext=(10,4),arrowprops=dict(facecolor='black',shrink=0.05))
ax.annotate('local manimum',xy=(5*np.pi,-1),xytext=(10,-6),arrowprops=dict(arrowstyle='->',connectionstyle="angle3,angleA=0,angleB=-90"))

plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(10,2,2000)
x=z*np.sin(30*z)
y=z*np.cos(30*z)

ax.plot3D(x,y,z,'green')
ax.set_title('3D line Plot')
plt.show()
```
