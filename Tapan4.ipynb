```python
import pandas as pd

s1=pd.Series([10,20,30,40,50])
print("The Series Value are: ",s1.values)
print("The index values are:", s1.index.values)
print(s1)
```

```python
s2=pd.Series([80,90,100,101,202],index=['English','Maths',"Science","EG",'Social Science'])
print(s2)
print("\n",s2.values)
print("\n",s2.index.values)
print("\n",s2['Social Science'])
print("\nMax:",s2.max())
print("\nMin:",s2.min())
print("\nCount:",s2.count())
print(f"\nEnglish: {s2['English']} EG: {s2['Maths']}",)
s2[1:2]
```
